    1295. Find Numbers with Even Number of Digits

```python
class Solution:
    def findNumbers(self, nums: List[int]) -> int:
        count=0
        for i in nums:
            if(len(str(i)))%2==0:
                count+=1
        return count
obj=Solution()
print(obj.findNumbers([54,46,546,5446]))
```

    1512. Number of Good Pairs

```python
class Solution:
    def numIdenticalPairs(self, nums: List[int]) -> int:
        freq = {}
        c = 0
        
        for i in nums:
            if i in freq:
                c += freq[i]
                freq[i] += 1
            else:
                freq[i] = 1
        
        return c
obj=Solution()
print(obj.numIdenticalPairs([1,2,3,1,1,3]))
```

    1365. How Many Numbers Are Smaller Than the Current Number

```python
class Solution:
    def smallerNumbersThanCurrent(self, nums: List[int]) -> List[int]:
        sorted_nums = sorted(nums)
        result = []
        for i in nums:
            result.append(sorted_nums.index(i))
        return result
obj=Solution()
print(obj.smallerNumbersThanCurrent([8,1,2,2,3]))
```

    1281. Subtract the Product and Sum of Digits of an Integer

```python
    def subtractProductAndSum(self, n: int) -> int:
        dummy=n
        temp1=1
        temp2=0
        for i in range(1,10000):
            if n>0:
                temp1*=n%10
                n=n//10
                temp2+=dummy%10
                dummy=dummy//10
        res=temp1-temp2
        return res
    print(subtractProductAndSum(1,234))
                
```

    1486. XOR Operation in an Array

```python
class Solution:
    def xorOperation(self, n: int, start: int) -> int:
        result = 0
        for i in range(n):
            result ^= start + 2 * i
        return result
hago=Solution()
print(hago.xorOperation(5,0))
```
