2. Python If-Else

```python
n=int(input())
if n%2!=0:
    print("Weird")
elif n%2==0 and n>=2 and n<=5:
    print("Not Weird")
elif n%2==0 and n>=6 and n<=20:
    print("Weird")
elif n%2==0 and n>20:
    print("Not Weird")
```

6. Write a function

```python
def is_leap(year):
    leap = False
    
    # Write your logic here
    if (year%4==0 and year%100!=0) or (year%400==0):
        leap=True
    else:
        leap=False
    
    return leap

year = int(input())
print(is_leap(year))
```
